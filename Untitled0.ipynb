```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
df = pd.read_csv("/content/drive/MyDrive/Ml_models/employee.csv")
df
reg = linear_model.LinearRegression()
reg.fit(df[['Salary']], df.ID)
%matplotlib inline
plt.xlabel('Salary')
plt.ylabel('ID')
plt.scatter(df.Salary,df.ID,color= "green",marker="+")
plt.plot(df.Salary, reg.predict(df[['Salary']]),color='blue')
```
